```python
# Ordinal - data is categorical but its comparable. like review (good, better, best), marks (distinction, fc, , 
# Nomial - data is categorical but not related to each other. (state, colour)
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ordinal_data.csv')
```

```python
df.head()
```

```python
df = df.iloc[:, 2:]
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:2], df.iloc[:, 2], test_size = 0.2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# CREATE OR OBJECT
```

```python
oe = OrdinalEncoder(categories = [['Poor', 'Average', 'Good'], ['School', 'UG','PG']])   # GIVE ORDER IN CATEGORIES ELSE IT WILL TAKE AUTOMATICALLY
```

```python
oe.fit(X_train)
```

```python
# Transform data

X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
