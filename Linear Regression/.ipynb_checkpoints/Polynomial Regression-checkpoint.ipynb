```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression, SGDRegressor

from sklearn.preprocessing import PolynomialFeatures, StandardScaler

from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 1.3 * x**2 + 0.4 * x + 7 + np.random.randn(200, 1)

# y = 1.3 x2 + 0.4x + 7   + (noise for random data)
```

```python
# visualize graph
plt.plot(x, y, '.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Why Linear/Multiple Regression won't work!

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 2, test_size = 0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
## VISUALIZE WHY LINEAR WON't WORK IN THIS CASE
plt.plot(x_train, lr.predict(x_train), color = 'red')
plt.plot(x, y, '.b')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot
plt.show()
```

### Polynomial Regression

```python
poly = PolynomialFeatures(degree = 2, include_bias = False)   #default True

x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)
```

```python
print(x_test_trans[0])
```

```python
# include bias
poly = PolynomialFeatures(degree = 2, include_bias = True)

x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)
```

```python
print(x_test_trans[0])
```

```python
## apply linear Regression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train_trans, y_train)
```

```python
y_pred = lr.predict(x_test_trans)
```

```python
r2_score(y_test, y_pred)  ## SEE HOW MUCH DATA IS COVERED
```

```python
# check formula y = 1.3 x2 + 0.4x + 7
print(lr.coef_)
print(lr.intercept_)
```

```python
x_new = np.linspace(-3, 3, 200).reshape(200, 1) 
x_new_poly = poly.transform(x_new)
y_new = lr.predict(x_new_poly)
```

```python
plt.plot(x_new, y_new, 'r-', label='Predictions')
plt.plot(x_train, y_train, 'b.', label = 'Training Data')
plt.plot(x_test, y_test, 'g.', label = 'Test Data')
plt.legend()
plt.show()
```
