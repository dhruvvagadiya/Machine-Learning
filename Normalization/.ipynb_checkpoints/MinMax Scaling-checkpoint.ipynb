```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine_data.csv', header = None).iloc[:, :3]
df.head()
```

```python
df.info()
```

```python
df.columns = ['Class Label', 'Alcohol', 'Malic Acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
sns.scatterplot(df, x = 'Alcohol', y = 'Malic Acid', hue = 'Class Label')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class Label', axis = 1), df['Class Label'], test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
## FIT DATA

scaler.fit(X_train)

X_train_trans = scaler.transform(X_train)
X_test_trans = scaler.transform(X_test)
```

```python
X_train_trans = pd.DataFrame(X_train_trans, columns = X_train.columns)
X_train_test = pd.DataFrame(X_test_trans, columns = X_train.columns)
```

```python
np.round(X_train.describe(), 1) 
```

```python
np.round(X_train_trans.describe(), 1) 
```

```python
## SEE DATA TRANSFORMED BW 0 TO 1
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.scatter(X_train['Alcohol'], X_train['Malic Acid'], c = y_train)
ax1.set_title('Before')

ax2.scatter(X_train_trans['Alcohol'], X_train_trans['Malic Acid'], c = y_train)
ax2.set_title('After')
```

```python
## NOW CHECK DISTRBUTION
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title('Before')
sns.kdeplot(X_train['Alcohol'], ax = ax1)
sns.kdeplot(X_train['Malic Acid'], ax = ax1)

ax2.set_title('After')
sns.kdeplot(X_train_trans['Alcohol'], ax = ax2)
sns.kdeplot(X_train_trans['Malic Acid'], ax = ax2)
```

```python
# BASICALLY EVERYTHING COMES DOWN BW 0 -- 1
```

```python
# CHECK INDIVIDUAL SCALINGS
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title('Alcohol before')
sns.kdeplot(X_train['Alcohol'], ax = ax1)

ax2.set_title('Alcohol after')
sns.kdeplot(X_train_trans['Alcohol'], ax = ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title('Malic Acid before')
sns.kdeplot(X_train['Malic Acid'], ax = ax1)

ax2.set_title('Malic Acid after')
sns.kdeplot(X_train_trans['Malic Acid'], ax = ax2)
```
